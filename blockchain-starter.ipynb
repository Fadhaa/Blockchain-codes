```python
import warnings
warnings.filterwarnings("ignore")
import urllib3
import gc
from time import time,sleep
```

```python
def gettrans(iz):
    htt=urllib3.PoolManager()
    ml = htt.request('GET', 'https://chain.api.btc.com/v3/tx/unconfirmed')
    ml=ml.data
    da=str(ml)
    da=da[-13800:]
    newda=da[(da.index(',"')+1):da.index(']},"err_code"')]
    fn=open('/kaggle/working/tx'+str(iz)+'.txt','w')
    fn.write(newda)
    fn.close()
    return newda
```

### To get the last N of unconfirmed latest transactions within blockchain

```python
gettrans(20)###to get latest 20 unconfirmed transactions 
```

## The code below to get information of the latest BITCOIN block added to the blockchain

```python
def getlatest():
	htt=urllib3.PoolManager()
	ml = htt.request('GET', 'https://chain.api.btc.com/v3/block/latest')
	ml=ml.data
	da=str(ml)
	fn=open('/kaggle/working/late.txt','w')
	fn.write(da)
	fn.close()
	return da
```
